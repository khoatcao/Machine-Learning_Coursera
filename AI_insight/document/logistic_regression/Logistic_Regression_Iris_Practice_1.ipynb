```python
inds = np.arange(X.shape[0]) # 100 rows 
np.random.shuffle(inds)
inds
```

```python
X = X[inds]
X
```

```python
y = y[inds]
y
```

```python
def sigmoid_function(z) : 
    return 1/(1+np.exp(-z)) # sigmoid function 
```

```python
def loss_function(h,y) :
    return (-y*np.log(h) - (1-y)*np.log(1-h)).mean() # mean for n sample
```

```python
def predict(X,theta) : # put z into sigmoid function 
    y_hat = sigmoid_function(np.dot(X,theta))
    return y_hat
```

```python
def compute_gradient(X,y,h): 
    return np.dot(X.T,(h-y))/y.size
```

```python
def update_theta(theta,gradient,learning_rate) : 
    return theta - learning_rate*gradient
```

```python
learning_rate = 0.1
epoch_max = 700 

theta = np.array([0.1,0.5,-0.1])
losses = []
accs = []
```

```python
# for training 
for epoch in range(epoch_max) : # iterate 700 epochs
    # predict z 
    y_hat = predict(X,theta) 
   # print("Predict",y_hat)
    # loss function 
    loss = loss_function(y_hat,y)
    #losses.append(loss)
    #print("Losses",losses)
    # gradient 
    gradient = compute_gradient(X,y,y_hat) 
    # update weights b and w
    theta = update_theta(theta,gradient,learning_rate)
    #print("Update theta",theta)
    ############# #############
    if(epoch%1 == 0) : # a v dng 0 v 1 
        # loss 
        losses.append(loss)
        #  accuracy for training
        preds = predict(X,theta).round()
        acc = (preds == y).mean()
        accs.append(acc)
        
        
        print("preds",preds)
        print("accs",np.mean(accs)*100.0)
```

```python
plt.plot(losses,label = "losses")
plt.ylabel("losses")
plt.xlabel("iteration")
```
